Hay multitud de factores y malas prácticas que llevan al sobre-ajuste, la más obvia es la del uso de ejemplos de entreno para posteriormente usarlos en la fase de test. 

El problema es el que sistema ya ha visto esos ejemplos en la fase de entreno y lo más probable es que tenga mucha facilidad para predecir los resultados si le volvemos a dar los mismos datos en la fase de test. Por este motivo no tenemos forma de evaluar correctamente o de saber si el modelo final realmente generaliza bien con ejemplos que no ha visto nunca, aunque es muy probable que haya overfitting si hay relativamente pocos datos, ya que para el modelo es bastante sencillo formar una línea o una curba que pase por cada uno de los puntos (en caso de una regresion lineal/polinómica) o formar una o varias que los separe (en un problema de clasificación). 

En el caso de overfitting la mejor opción es añadir datos "nuevos" al set de entreno con cualquier técnica de aumento de datos y como segunda opcion usar algún metodo de regularizacion (L1, L2 o Dropout (entre otros)) con el mismo set y posteriormente usar un segundo y tercer set de validación y test respectivamente, para intentar mejorar el modelo y testearlo con ejemplos que nunca haya visto

En los siguientes ejercicios de la PEC usaremos el conjunto de datos del dataset `breast_cancer` de *sklearn*. Crea una tabla con los estadísticos descriptivos de todas las características de los tumores de la base de datos `breast_cancer` agrupados por tipo (benigno/maligno).

Realizad un test estadístico [ttest](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html) para comparar el error de la simetría (*symmetry error*) y tersura (*mean smoothness*) de los tumores benignos con las de los malignos. Dado el resultado obtenido, indica qué características podrían ser más relevantes para diagnosticar un tumor maligno.

Nota: Consulta este [enlace](http://www.conexionismo.com/leer_articulo.php?ref=prueba_t_de_student_para_la_comparacion_de_dos_muestras_independientes-j960497l) si quieres ampliar conocimientos sobre el test estadístico ttest. Para realizar el test estadístico automáticamente podéis utilizar el método [`ttest_ind`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html) de *Scipy.stats*
**(1 punto)** <span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span>


```python
# Respuesta
#Repaso libros del CFA, bloque de "quantitative methods"

#Importamos librerias
import pandas as pd
from sklearn import datasets
from scipy import stats

#Cargamos el dataset y creamos un dataframe de pandas con una columna nueva "tumor type" para tumores malignos/benignos
brst = datasets.load_breast_cancer()
df = pd.DataFrame(brst.data, columns = brst.feature_names)
df['tumor type'] = brst['target_names'][brst['target']]

#Agrupamos por tipo de tumor y creamos la tabla "stat"
group = df.groupby(['tumor type'])
stat = group.describe()
#print(stat)

#Estudiamos si las medias de ambos atributos respecto a tumores malignos y benignos son estadísticamente 
#diferentes con el T-Test
var1 = df.where(df['tumor type'] == 'benign').dropna()
var2 = df.where(df['tumor type'] == 'malignant').dropna()
zval_sym, pval_sym = stats.ttest_ind(var1['symmetry error'], var2['symmetry error'])
zval_smo, pval_smo = stats.ttest_ind(var1['mean smoothness'], var2['mean smoothness'])


print('''El P-value {} es demasiado grande para rechazar la hipótesis 
de que la media del error de simetría es la misma en la distribución de tumores benignos y malignos, por lo tanto
es la misma en ambas. El Z-value es de sólo {} desviaciones estándar respecto a la media, donde 
hay más densidad de valores cercanos a la media\n\n'''.format(pval_sym,zval_sym))

print('''Con un P-value o nivel de significancia del {} o 0.00000000000000000105 podemos rechazar 
la hipótesis de que la media de la tersura es la misma en la distribución de tumores benignos y malignos,
por lo tanto es diferente en ambas. El Z-value es de {} desviaciones estándar respecto a la media, donde 
apenas hay valores, en la parte mas estrecha de la cola izquierda'''.format(pval_smo,zval_smo))
```

#### Respuesta

En el caso del error de simetría en tumores benignos y malignos vemos que no hay una diferencia estadísticamente significativa para decir que son diferentes, ya que el valor mínimo de significancia para rechazar la hipótesis nula (P-value) se encuentra muy cerca de la media de la distribución. Aunque éste valor fuera un 0.05 (0.025 cola izquierda + 0.025 cola derecha) en lugar de 0.87 todavía podríamos decir que se trata de un valor que encontramos debajo de la campana el 95% de las veces, a ~1.96 desviaciones estándar, y no considerar las medias de las distribuciones de tumores benignos y malignos estadísticamente diferentes. Al contrario que ocurre si analizamos la tersura entre ambos tipos de tumor. En este caso podemos afirmar con un 99.9999999999999999895% de confianza (1-P-value) que la media de la tersura en la distribucion de tumores benignos y malignos es diferente. Se trata de un atributo importante para decidir si un tumor es maligno o benigno

## Ejercicio 4

El objetivo de un modelo de **regresión lineal** es encontrar una relación entre una o más características (variables independientes) y una variable objetivo continua (variable dependiente). Cuando sólo usamos una característica predictiva se le llama **Regresión Lineal Univariada** y si hay múltiples predictores se llama **Regresión Lineal Múltiple**.

Cuando la variable dependiente es una variable binaria que contiene datos codificados como 1 (sí, éxito, etc.) o 0 (no, fallo, etc.), como es el caso del dataset `breast_cancer`, tendremos que emplear una [regresión logística](https://ca.wikipedia.org/wiki/Regressi%C3%B3_log%C3%ADstica).

Crea un modelo de regresión logística que estime la probabilidad de que un tumor sea maligno dada su área y su textura. Muestra la [matriz de confusión](https://ca.wikipedia.org/wiki/Matriu_de_confusi%C3%B3) del modelo obtenido.

Nota: Quizás te interesa explorar la función [sklearn.metrics.confusion_matrix()](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html) **(1,5 puntos)** <span style="font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px; ">EI</span>

```python
# Respuesta
#https://towardsdatascience.com/logistic-regression-using-python-sklearn-numpy-mnist-handwriting-recognition-matplotlib-a6b31e2b166a
#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

#Importamos librerias
import warnings
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


#Cargamos el set, nos quedamos con los atributos área y textura y normalizamos
brst = datasets.load_breast_cancer()
X = brst.data[:,(1,3)]
labels = brst.target
sc = StandardScaler()
X = sc.fit_transform(X)

# Separamos las muestras utilizando un 30 % para test y el resto para aprendizaje.
X_train, X_test, Y_train, Y_test = train_test_split(X, labels, test_size=0.3, random_state=0)

#Importamos el regresor y lo entrenamos con todo el batch
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)

#Predicciones
warnings.filterwarnings('ignore')
pred = log_reg.predict(X_test)

#Matriz de confusion con True Negative (en este caso el 0 es maligno) [0,0],
#False Positive [0,1], False Negative[1,0] y True Positive [1,1]
cm = confusion_matrix(Y_test, pred)
print("Matriz de confusion:\n\n", cm)
```

```python
#% Aciertos:
aciertos_totales = (cm[0,0] + cm[1,1])/ (cm[0,0] + cm[0,1] + cm[1,0] + cm[1,1])
print('% Aciertos totales: ', aciertos_totales)

#Con esto podemos encontrar la precisión, es decir, de los que hemos pronosticado con la clase menos frecuente,
#en este caso los negativos (malignos) cuantos realmente lo son:
precision = cm[0,0] / (cm[0,0] + cm[1,0]) 
print('Precision:', precision)

#Y el "recall", es decir, de los que son realmente negativos (malignos), cuántos hemos detectado que lo son
recall = cm[0,0] / (cm[0,0] + cm[0,1])
print('Recall:   ', recall)

#Y una métrica como el F1 score (o media harmónica) para poder escoger de una forma más objetiva entre modelos 
#de los cuales no estamos seguros cuál es el mejor. La media harmónica penaliza cuando un modelo no tiene un equilibrio 
#entre el recall y la precision (es decir, si el numerador tiene a cero)
f1_score = 2 * (precision * recall)/(precision + recall)
print('F1 Score: ', f1_score)
```

## Ejercicio 5

El **Multi-layer Perceptron (MLP)** es un algoritmo de aprendizaje supervisado que aprende una función mediante el entrenamiento de un modelo que asocia n dimensiones de entrada (*predictores*) con dimensiones de salida (*targets*). Teniendo en cuenta un conjunto de funciones y un objetivo, puede aprender un aproximador de funciones no lineales por clasificación o regresión. Es diferente de la regresión logística, ya que entre la capa de entrada y la de salida, puede haber una o más capas no lineales, llamadas capas ocultas.

Aplica un clasificador basado en un MLP para predecir los tipos de tumor utilizando el área, la textura, la simetría y la concavidad, como atributos y utilizando el 70% de las muestras de entrenamiento y el 30% de test. Debéis utilizar la función [sklearn.neural_network.MLPClassifier](https://scikit-learn.org/stable/modules/neural_networks_supervised.html).

Utiliza tres niveles (*layers*) con 30 neuronas por nivel. ¿Qué valor de precisión obtenemos en un modelo basado en un MLP? **(2 puntos)** <span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span>

```python
# Respuesta
#https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

#Importamos librerias
import warnings
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier


#Cargamos el set, nos quedamos con los atributos área, textura, simetría y concavidad y normalizamos
brst = datasets.load_breast_cancer()
X = brst.data[:,(1,3,6,8)]
labels = brst.target
sc = StandardScaler()
X = sc.fit_transform(X)

# Separamos las muestras utilizando un 30 % para test y el resto para aprendizaje.
X_train, X_test, Y_train, Y_test = train_test_split(X, labels, test_size=0.3, random_state=0)

#Importamos el regresor y lo entrenamos con todo el batch (569*0.7 = ~390)
ful_con = MLPClassifier(hidden_layer_sizes=(30,30,30),learning_rate_init=0.001,
                        activation='relu', batch_size=512, shuffle=True)
ful_con.fit(X_train, Y_train)

#Predicciones
warnings.filterwarnings('ignore')
pred = ful_con.predict(X_test)

#Matriz de confusion con True Negative (en este caso el 0 es maligno) [0,0],
#False Positive [0,1], False Negative[1,0] y True Positive [1,1]
cm = confusion_matrix(Y_test, pred)
print("Matriz de confusion:\n\n", cm)
```

```python
#% Aciertos:
aciertos_totales = (cm[0,0] + cm[1,1])/ (cm[0,0] + cm[0,1] + cm[1,0] + cm[1,1])
print('\n\n% Aciertos totales: ', aciertos_totales)

#Con esto podemos encontrar la precisión, es decir, de los que hemos pronosticado con la clase menos frecuente,
#en este caso los negativos (0, malignos) cuántos realmente lo son:
precision = cm[0,0] / (cm[0,0] + cm[1,0]) 
print('Precision:', precision)

#Y el "recall", es decir, de los que son realmente negativos (malignos), cuántos hemos detectado que lo son
recall = cm[0,0] / (cm[0,0] + cm[0,1])
print('Recall:   ', recall)

#Y una métrica como el F1 score (o media harmónica) para poder escoger de una forma más objetiva entre modelos 
#de los cuales no estamos seguros cuál es el mejor. La media harmónica penaliza cuando un modelo no tiene un equilibrio 
#entre el recall y la precision (es decir, si el numerador tiene a cero)
f1_score = 2 * (precision * recall)/(precision + recall)
print('F1 Score: ', f1_score)
```

Con todo lo demás igual y añadiendo 2 atributos más tenemos un F1 Score más alto que con la regresión logística. Es coherente ya que la regresión logística solo tiene la capa de entrada y la neurona de salida, por lo que no puede aprender funciones demasiado complejas. Sin embargo una red neuronal puede aprender funciones más interesantes con las capas intermedias (sobretodo si añadimos mas neuronas). De todos modos, para un problema con tan pocos datos usar la regresión logística se puede considerar más que aceptable

Carga el conjunto de datos breast_cancer incorporado en la librería `sklearn`. Implementa una función, `descriu_diagnostics`, que devuelva un diccionario con la siguiente estructura:

`` 
{
   "Categorias": [],
   "Atributos": [],
   "Num_muestras": 0
}
`` 

*Categorias* debe ser un array con el nombre de los **targets** del dataset. *Atributos* debe ser un array con el nombre de los **atributos** y finalmente, *num_muestras* debe indicar el número **total de muestras** del dataset.

```python
from sklearn import datasets

# Carga el dataset breast_cancer

diagnostics = datasets.load_breast_cancer()

def describe_diagnostics():
    #Inicialitzem el diccionari
    diccionario = {}
    #Assignem a cada atribut el valor corresponent segons al dataset breast_cancer
    diccionario["categorias"] = diagnostics.target_names
    diccionario["atributos"] = diagnostics.feature_names
    diccionario["num_muestras"] = len(diagnostics.data)
    
    return diccionario

print(describe_diagnostics())
```
